```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import math
sns.set(style="darkgrid")

df = pd.read_csv('2020-XTern-DS.csv')
df.columns = ['id', 'lat', 'long', 'food', 'avg_cost', 'min_order', 'rate', 'votes', 'review', 'time']
df.time = df.time.str.extract("(\d*\.?\d+)", expand=True)
df.review = df.review.str.extract("(\d*\.?\d+)", expand=True)
df.votes = df.votes.str.extract("(\d*\.?\d+)", expand=True)
df.rate = df.rate.str.extract("(\d*\.?\d+)", expand=True)
df.min_order = df.min_order.str.extract("(\d*\.?\d+)", expand=True)
df.avg_cost = df.avg_cost.str.extract("(\d*\.?\d+)", expand=True)
df.time = df.time.apply(pd.to_numeric)
df.review = df.review.apply(pd.to_numeric)
df.votes = df.votes.apply(pd.to_numeric)
df.rate = df.rate.apply(pd.to_numeric)
df.min_order = df.min_order.apply(pd.to_numeric)
df.avg_cost = df.avg_cost.apply(pd.to_numeric)
df
```

```python
foodToVotes = {}
for ind in df.index: 
    if(math.isnan(df.votes[ind])) :
        continue;
    else :
        foodTypes = df.food[ind].split(", ")
        for food in foodTypes :
            if food in foodToVotes :
                foodToVotes[food] += df.votes[ind]
            else :
                foodToVotes[food] = df.votes[ind]
            
votesToFood = dict([(value, key) for key, value in foodToVotes.items()]) 
sortDict = sorted(votesToFood, reverse=True)
print("The top 5 resturant types by user voting are:\n")
print("1: ", votesToFood[sortDict[0]] + " with " + str(sortDict[0]) + " votes")
print("2: ", votesToFood[sortDict[1]] + " with " + str(sortDict[1]) + " votes")
print("3: ", votesToFood[sortDict[2]] + " with " + str(sortDict[2]) + " votes")
print("4: ", votesToFood[sortDict[3]] + " with " + str(sortDict[3]) + " votes")
print("5: ", votesToFood[sortDict[4]] + " with " + str(sortDict[4]) + " votes")
print("\n\n")

sortDict = sorted(votesToFood)
print("The worst 5 resturant types by user voting are:\n")
print("1: ", votesToFood[sortDict[0]] + " with " + str(sortDict[0]) + " votes")
print("2: ", votesToFood[sortDict[1]] + " with " + str(sortDict[1]) + " votes")
print("3: ", votesToFood[sortDict[2]] + " with " + str(sortDict[2]) + " votes")
print("4: ", votesToFood[sortDict[3]] + " with " + str(sortDict[3]) + " votes")
print("5: ", votesToFood[sortDict[4]] + " with " + str(sortDict[4]) + " votes")
```

```python
foodToRating = {}
for ind in df.index: 
    if(math.isnan(df.rate[ind])) :
        continue;
    else :
        foodTypes = df.food[ind].split(", ")
        for food in foodTypes :
            if food in foodToRating :
                foodToRating[food] += df.rate[ind]
            else :
                foodToRating[food] = df.rate[ind]
            
ratingToFood = dict([(value, key) for key, value in foodToRating.items()]) 
sortDict = sorted(ratingToFood, reverse=True)
print("The top 5 resturant types by user rating are:\n")
print("1: ", ratingToFood[sortDict[0]] + " with a cumulitive rating of " + str(round(sortDict[0])))
print("2: ", ratingToFood[sortDict[1]] + " with a cumulitive rating of " + str(round(sortDict[1])))
print("3: ", ratingToFood[sortDict[2]] + " with a cumulitive rating of " + str(round(sortDict[2])))
print("4: ", ratingToFood[sortDict[3]] + " with a cumulitive rating of " + str(round(sortDict[3])))
print("5: ", ratingToFood[sortDict[4]] + " with a cumulitive rating of " + str(round(sortDict[4])))
print("\n\n")

sortDict = sorted(ratingToFood)
print("The worst 5 resturant types by user rating are:\n")
print("1: ", ratingToFood[sortDict[0]] + " with a cumulitive rating of " + str(round(sortDict[0])))
print("2: ", ratingToFood[sortDict[1]] + " with a cumulitive rating of " + str(round(sortDict[1])))
print("3: ", ratingToFood[sortDict[2]] + " with a cumulitive rating of " + str(round(sortDict[2])))
print("4: ", ratingToFood[sortDict[3]] + " with a cumulitive rating of " + str(round(sortDict[3])))
print("5: ", ratingToFood[sortDict[4]] + " with a cumulitive rating of " + str(round(sortDict[4])))
```

```python
# Most popular resturant = (rating*votes)/(avg_cost*cookTime)
foodToScore = {}
for ind in df.index: 
    score = df.rate[ind]*df.votes[ind]/(df.avg_cost[ind]*df.time[ind])
    if (math.isnan(score)) :
       continue;
    else :
        foodToScore[df.id[ind]] = score

scoreToFood = dict([(value, key) for key, value in foodToScore.items()]) 
sortDict = sorted(scoreToFood, reverse=True)
print("The top 5 resturant types by overall score are:\n")
print("1: ", scoreToFood[sortDict[0]] + " with a score of " + str(round(sortDict[0])))
print("2: ", scoreToFood[sortDict[1]] + " with a score of " + str(round(sortDict[1])))
print("3: ", scoreToFood[sortDict[2]] + " with a score of " + str(round(sortDict[2])))
print("4: ", scoreToFood[sortDict[3]] + " with a score of " + str(round(sortDict[3])))
print("5: ", scoreToFood[sortDict[4]] + " with a score of " + str(round(sortDict[4])))

print("\n\n")
sortDict = sorted(scoreToFood)
print("The worst 5 resturant types by overall score are:\n")
print("1: ", scoreToFood[sortDict[0]] + " with a score of " + str(round(sortDict[0])))
print("2: ", scoreToFood[sortDict[1]] + " with a score of " + str(round(sortDict[1])))
print("3: ", scoreToFood[sortDict[2]] + " with a score of " + str(round(sortDict[2])))
print("4: ", scoreToFood[sortDict[3]] + " with a score of " + str(round(sortDict[3])))
print("5: ", scoreToFood[sortDict[4]] + " with a score of " + str(round(sortDict[4])))
```
